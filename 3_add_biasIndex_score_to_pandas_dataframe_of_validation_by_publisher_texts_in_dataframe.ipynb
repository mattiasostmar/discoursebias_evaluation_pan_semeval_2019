```python
import pandas as pd
import sys
sys.path.append('../discoursebias/')
import discoursebias as dib
```

```python
df = pd.read_pickle("article_validation_bypublisher_20181122.pickle")
```

```python
df.head(3)
```

```python
df.text.str.contains("\n\n")
```

```python
bis = []
for ix, row in df.iterrows():
    exp = dib.TextNetwork(row["text"], textname=row["id"], plot=False, save_gexf=False)
    res = exp.run()
    bis.append(res.loc["biasIndex"].values[0])
df["biasIndex"] = pd.Series(bis)
```
